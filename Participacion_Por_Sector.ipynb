```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


# Datos del PIB millones de pesos y por participaciones porcentuales
datos = {
    'Año': [2019, 2020, 2021, 2022, 2023],
    'PIB Informal (%)': [23.3, 24.0, 24.4, 24.8, 24.8],
    'PIB Formal (%)': [76.7, 76.0, 75.6, 75.2, 75.2]
}

# Creacion del DataFrame
df = pd.DataFrame(datos)

# Posicionamiento de Barras
x = np.arange(len(df['Año']))
width = 0.4

# Gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, df['PIB Informal (%)'], width, color='orange', label='Sector Informal')
plt.bar(x + width/2, df['PIB Formal (%)'], width, color='purple', label='Sector Formal')
plt.xticks(x, df['Año'])
plt.title('Participación del PIB por Sector (Gráfico de Barras)')
plt.xlabel('Año')
plt.ylabel('Participación (%)')
plt.ylim(0, 100)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Datos Extraídos del INEGI en https://www.inegi.org.mx/contenidos/saladeprensa/boletines/2024/MDEI/MDEI2023.pdf
plt.figtext(0.5, -0.05, 'Fuente: INEGI, elaboración propia', ha='center', fontsize=10)
plt.tight_layout()
plt.show()

#By: Pineda Román Jaime Chrystopher
```
