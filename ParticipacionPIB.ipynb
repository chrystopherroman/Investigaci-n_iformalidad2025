```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos oficiales del INEGI en 2023 Proyecto Investigacin Chrystopher Pineda
data = {
    'Ao': [2019, 2020, 2021, 2022, 2023],
    'PIB_Formal': [82.3, 79.1, 80.5, 81.7, 82.1],  # % del PIB
    'PIB_Informal': [17.7, 20.9, 19.5, 18.3, 17.9]  # % del PIB
}

df = pd.DataFrame(data)

plt.style.use('seaborn-v0_8')
plt.figure(figsize=(10, 6))

plt.bar(df['Ao'], df['PIB_Formal'], color='#8e44ad', label='Sector Formal')     # Morado
plt.bar(df['Ao'], df['PIB_Informal'], bottom=df['PIB_Formal'], color='#e67e22', label='Sector Informal')  # Naranja

plt.title('Participacin en el PIB de Mxico: Sector Formal vs Informal (2019-2023)',
          fontsize=14, fontweight='bold')
plt.xlabel('Ao', fontsize=12)
plt.ylabel('Porcentaje del PIB (%)', fontsize=12)
plt.ylim(0, 100)
plt.xticks(df['Ao'])

for i, year in enumerate(df['Ao']):
    plt.text(year, df['PIB_Formal'][i]/2, f"{df['PIB_Formal'][i]}%",
             ha='center', va='center', color='white', fontweight='bold')
    plt.text(year, df['PIB_Formal'][i] + df['PIB_Informal'][i]/2, f"{df['PIB_Informal'][i]}%",
             ha='center', va='center', color='white', fontweight='bold')

plt.legend(loc='upper right')
plt.figtext(0.5, 0.01, "Elaboracin propia. Datos extrados del INEGI.",
            ha="center", fontsize=10, style='italic')

plt.tight_layout()
plt.show()

```
